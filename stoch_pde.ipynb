```python
# ou process
ou = pd.read_csv("ou_process.txt")
ou
```

```python
x = np.linspace(-.15,.15,100)
y = (1/np.sqrt(np.pi*0.07*0.07))*np.exp(-x**2/(0.07*0.07))
```

```python
# equilibrium for ornstein uhlenbeck process
plt.plot(x,y)
plt.hist(ou,density=True,bins=20,alpha=0.5)
plt.show()
```

```python
ou.plot(figsize=(15,7))
plt.show()
```

```python
corr_func = []
for i in range(0,1000):
    corr_func.append((ou.shift(i) * ou).mean())
```

```python
pd.DataFrame(corr_func).plot() # exponential correlation function
```
